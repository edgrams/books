return only fields interested in

```bash
curl -XGET 'localhost:9200/_search?pretty' -H 'Content-Type: application/json' -d'
{
    "query":   { "match_all": {}},
    "_source": [ "tweet" ]
}
'
```

match the all field

```bash
curl -XGET 'localhost:9200/_search?pretty' -H 'Content-Type: application/json' -d'
{
    "query": {"match": {"_all": "john smith marketing"} }
}
'
```

disable _all field

```bash
curl -XPUT 'localhost:9200/my_index/_mapping/my_type' -H 'Content-Type: application/json' -d'
{
    "my_type": {
        "_all": { "enabled": false }
    }
}
'
```

include only certain fields in _all

```bash
curl -XPUT 'localhost:9200/my_index/_mapping/my_type' -H 'Content-Type: application/json' -d'
{
    "my_type": {
        "include_in_all": false,
        "properties": {
            "title": {
                "type":           "string",
                "include_in_all": true
            }
        }
    }
}
'
```

configure analyzer for _all field

```bash
curl -XPUT 'localhost:9200/my_index/_mapping/my_type' -H 'Content-Type: application/json' -d'
{
    "my_type": {
        "_all": { "analyzer": "whitespace" }
    }
}
'
```
